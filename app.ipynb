````python
import json

# Step 1: read contact data
with open("contacts.json", "r", encoding="utf-8") as file:
    contacts = json.load(file)

print("All contacts:")
for contact in contacts:
    print(contact)

# Step 2: group contacts by persona
segments = {}

for contact in contacts:
    persona = contact["persona"]

    if persona not in segments:
        segments[persona] = []

    segments[persona].append(contact)

print("\nSegmented contacts:")
for persona, people in segments.items():
    print(f"\n=== {persona} ===")
    for person in people:
        print(person["email"])
````

````python
blog_data = {
    "title": "How AI Is Transforming Creative Agency Workflows",
    "outline": [
        "Why creative agencies are overloaded",
        "How AI reduces repetitive work",
        "Examples of AI in agency operations",
        "How teams can stay creative while using automation"
    ],
    "blog_draft": """Creative agencies are under pressure to do more with fewer resources. 
From managing client communication to coordinating internal workflows, teams often spend too much time on repetitive work that slows down creativity.

AI tools are helping agencies automate routine tasks such as scheduling, follow-up emails, project updates, and content planning. Instead of replacing creative teams, these tools can reduce operational overhead and free up more time for strategic and creative work.

For founders, AI can support scale without immediately increasing headcount. For operations managers, AI can reduce handoff issues and improve process visibility. For creative professionals, AI can remove admin-heavy tasks and make more room for actual creative output.

The most effective agencies will not be the ones that automate everything blindly. They will be the ones that use AI thoughtfully to improve speed, consistency, and team focus while protecting the human side of creative work."""
}

newsletters = {
    "founder": {
        "subject_line": "Scale your agency without scaling headcount",
        "body": "See how AI can reduce repetitive work, improve efficiency, and help your agency grow without adding unnecessary overhead."
    },
    "ops_manager": {
        "subject_line": "Reduce workflow bottlenecks with AI",
        "body": "Explore how AI can improve team coordination, reduce manual tasks, and make agency operations run more smoothly."
    },
    "creative_pro": {
        "subject_line": "Let AI handle the admin work",
        "body": "Learn how AI can take repetitive tasks off your plate so you can spend more time on creative thinking and execution."
    }
}

print("Blog title:")
print(blog_data["title"])

print("\nNewsletter versions:")
for persona, content in newsletters.items():
    print(f"\n{persona}")
    print("Subject:", content["subject_line"])
    print("Body:", content["body"])
````

````python
print("Campaign preview:")

for persona, people in segments.items():
    print(f"\n=== Sending to {persona} ===")
    print("Subject:", newsletters[persona]["subject_line"])

    for person in people:
        print("Recipient:", person["email"])
````

````python
from datetime import datetime

campaign_logs = []

for persona, people in segments.items():
    campaign = {
        "blog_title": blog_data["title"],
        "persona": persona,
        "newsletter_subject": newsletters[persona]["subject_line"],
        "recipient_count": len(people),
        "send_date": datetime.now().isoformat()
    }
    campaign_logs.append(campaign)

print("Campaign logs:")
for log in campaign_logs:
    print(log)
````

````python
performance_data = [
    {
        "persona": "founder",
        "open_rate": 0.44,
        "click_rate": 0.13,
        "unsubscribe_rate": 0.01
    },
    {
        "persona": "ops_manager",
        "open_rate": 0.39,
        "click_rate": 0.09,
        "unsubscribe_rate": 0.008
    },
    {
        "persona": "creative_pro",
        "open_rate": 0.48,
        "click_rate": 0.16,
        "unsubscribe_rate": 0.012
    }
]

print("Performance data:")
for item in performance_data:
    print(item)
````

````python
best_persona = None
highest_click_rate = 0

for item in performance_data:
    if item["click_rate"] > highest_click_rate:
        highest_click_rate = item["click_rate"]
        best_persona = item["persona"]

print("Best performing segment:")
print(best_persona)
print("Highest click rate:")
print(highest_click_rate)
````

````python
summary = """
Creative professionals had the highest click rate at 16%, suggesting that creator-focused messaging resonated most strongly.
Founders also performed well, especially with growth and efficiency-focused messaging.
Operations managers showed lower click engagement, which may mean this segment needs more specific workflow examples or clearer operational value.
For the next campaign, NovaMind should test more case-study-based content for creative professionals and more process-driven messaging for operations managers.
"""

print("Performance summary:")
print(summary)
````

````python
output = {
    "blog": blog_data,
    "newsletters": newsletters,
    "campaign_logs": campaign_logs,
    "performance_data": performance_data
}

import json

with open("output.json", "w", encoding="utf-8") as file:
    json.dump(output, file, indent=2, ensure_ascii=False)

print("Saved to output.json!")
````

````python
from openai import OpenAI

client = OpenAI()

response = client.responses.create(
    model="gpt-4.1-mini",
    input="Write one short sentence about AI in marketing."
)

print(response.output_text)
````

````python
topic = "AI in creative automation"

blog_prompt = f"""
You are a B2B SaaS content marketer for NovaMind, a fictional startup that helps small creative agencies automate workflows.

Write content for this topic:
{topic}

Please generate:
1. A blog title
2. A blog outline with 4 short bullet points
3. A blog draft of about 400 to 600 words

Requirements:
- Audience: small creative agencies
- Tone: practical, modern, clear
- Focus on workflow automation, efficiency, and creativity
- Avoid overly generic buzzwords

Return valid JSON only with this exact structure:
{{
  "title": "string",
  "outline": ["point1", "point2", "point3", "point4"],
  "blog_draft": "string"
}}
"""

response_blog = client.responses.create(
    model="gpt-4.1-mini",
    input=blog_prompt
)

print(response_blog.output_text)
````

````python
import json

blog_data = json.loads(response_blog.output_text)

print("Blog title:")
print(blog_data["title"])

print("\nOutline:")
for point in blog_data["outline"]:
    print("-", point)

print("\nDraft preview:")
print(blog_data["blog_draft"][:500])
````

````python
newsletter_prompt = f"""
You are a B2B email marketer for NovaMind.

Based on this blog:

Title: {blog_data["title"]}

Blog draft:
{blog_data["blog_draft"]}

Create 3 short newsletter versions for these personas:
1. founder
2. ops_manager
3. creative_pro

Requirements:
- Each version must include:
  - subject_line
  - body
- Each body should be about 50 to 90 words
- Make each version feel personalized to that persona
- founder: focus on growth, scale, ROI
- ops_manager: focus on workflow, efficiency, coordination
- creative_pro: focus on reducing admin work and protecting creativity

Return valid JSON only with this exact structure:
{{
  "founder": {{
    "subject_line": "string",
    "body": "string"
  }},
  "ops_manager": {{
    "subject_line": "string",
    "body": "string"
  }},
  "creative_pro": {{
    "subject_line": "string",
    "body": "string"
  }}
}}
"""

response_newsletters = client.responses.create(
    model="gpt-4.1-mini",
    input=newsletter_prompt
)

print(response_newsletters.output_text)
````

````python
print(response_newsletters.output_text)
````

````python
raw_text = response_newsletters.output_text.strip()
raw_text = raw_text.replace("```json", "").replace("```", "").strip()

start = raw_text.find("{")
end = raw_text.rfind("}") + 1
clean_json = raw_text[start:end]

newsletters = json.loads(clean_json)

print("Newsletter versions:")

for persona, content in newsletters.items():
    print(f"\n=== {persona} ===")
    print("Subject:", content["subject_line"])
    print("Body:", content["body"])
````
